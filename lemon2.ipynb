**train/valid split 후, binary encoding / mean encoding**

```python
train_categories =  x_train['제조사'].unique()
```

```python
# 제조사 값 별로 정수 치환

COL = '제조사'

train_cats = x_train[COL].astype("string").str.strip().unique()

train_cat = pd.Categorical(x_train[COL].astype("string").str.strip(), categories=train_cats)
valid_cat = pd.Categorical(x_valid[COL].astype("string").str.strip(), categories=train_cats)


train_codes = train_cat.codes
valid_codes = valid_cat.codes
```

```python
# 정수를 바이너리 코드로 변경

import numpy as np
import pandas as pd

COL = "제조사"  # 원본 컬럼명

# 1) 비트 수 계산 (제조사 고유값 개수 기준)
n_unique = len(train_cats)   # 너가 이미 만든 train_cats 사용
n_bits = int(np.ceil(np.log2(n_unique)))  # 필요한 비트 수

# 2) train_codes / valid_codes -> 바이너리 컬럼 생성
train_bin = pd.DataFrame(
    {f"{COL}_b{i}": ((train_codes >> i) & 1) for i in range(n_bits)},
    index=x_train.index
)

valid_bin = pd.DataFrame(
    {f"{COL}_b{i}": ((valid_codes >> i) & 1) for i in range(n_bits)},
    index=x_valid.index
)

# 3) 원본 컬럼 제거 + 바이너리 컬럼 붙이기
x_train = pd.concat([x_train.drop(columns=[COL]), train_bin], axis=1)
x_valid = pd.concat([x_valid.drop(columns=[COL]), valid_bin], axis=1)
```


**차량크기-one-hot Encoding**

```python
import pandas as pd

COL3 = "차량크기"

def norm(s):
    return (
        s.astype("string")
         .str.strip()
         .str.replace(r"\s+", " ", regex=True)
    )

train_cats3 = norm(x_train[COL3]).unique()

x_train_cat3 = pd.Categorical(norm(x_train[COL3]), categories=train_cats3)
x_valid_cat3 = pd.Categorical(norm(x_valid[COL3]), categories=train_cats3)

train_oh = pd.get_dummies(x_train_cat3, prefix=COL3, dummy_na=True).astype("uint8")
valid_oh = pd.get_dummies(x_valid_cat3, prefix=COL3, dummy_na=True).astype("uint8")

# ★ 인덱스 정렬 핵심
train_oh.index = x_train.index
valid_oh.index = x_valid.index

x_train = pd.concat([x_train.drop(columns=[COL3]), train_oh], axis=1)
x_valid = pd.concat([x_valid.drop(columns=[COL3]), valid_oh], axis=1)

x_valid = x_valid.reindex(columns=x_train.columns, fill_value=0)

assert x_train.shape[0] == y_train.shape[0]
assert x_valid.shape[0] == y_valid.shape[0]
```

**구매지우편번호-Mean Encoding**

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import StratifiedKFold

COL2 = "구매지우편번호"

def norm_zip(s: pd.Series) -> pd.Series:
    return s.astype("string").str.strip()

def make_oof_target_encoding(X: pd.DataFrame, y: pd.Series, col: str,
                            m: float = 20.0, n_splits: int = 5, seed: int = 42) -> pd.Series:
    X = X.copy()
    y = y.copy()
    global_mean = y.mean()

    oof = pd.Series(index=X.index, dtype="float64", name=f"{col}_te")
    skf = StratifiedKFold(n_splits=n_splits, shuffle=True, random_state=seed)

    for fit_idx, hold_idx in skf.split(X, y):
        X_fit = X.iloc[fit_idx]
        y_fit = y.iloc[fit_idx]

        stats = y_fit.groupby(X_fit[col]).agg(["sum", "count"])
        enc_map = (stats["sum"] + m * global_mean) / (stats["count"] + m)

        oof.iloc[hold_idx] = (
            X.iloc[hold_idx][col]
            .map(enc_map)
            .fillna(global_mean)
            .astype(float)
        )

    return oof

# --- 안전하게 copy로 작업(원본 x_train/x_valid 꼬임 방지) ---
Xtr = x_train.copy()
Xva = x_valid.copy()
ytr = y_train.copy()

# 1) 컬럼 존재 체크 (여기서 False면 이미 drop했거나 이름이 다름)
if COL2 not in Xtr.columns or COL2 not in Xva.columns:
    raise KeyError(f"{COL2} 컬럼이 x_train/x_valid 중 하나에 없습니다. "
                   f"train_has={COL2 in Xtr.columns}, valid_has={COL2 in Xva.columns}")

# 2) 정규화
Xtr.loc[:, COL2] = norm_zip(Xtr[COL2])
Xva.loc[:, COL2] = norm_zip(Xva[COL2])

# 3) train OOF TE 생성
Xtr[f"{COL2}_te"] = make_oof_target_encoding(Xtr, ytr, COL2, m=20.0, n_splits=5)

# 4) valid TE 생성 (train 전체 map으로)
global_mean = ytr.mean()
stats_all = ytr.groupby(Xtr[COL2]).agg(["sum", "count"])
enc_map_all = (stats_all["sum"] + 20.0 * global_mean) / (stats_all["count"] + 20.0)

Xva[f"{COL2}_te"] = Xva[COL2].map(enc_map_all).fillna(global_mean).astype(float)

# 5) 원본 우편번호 drop (둘 다)
Xtr = Xtr.drop(columns=[COL2])
Xva = Xva.drop(columns=[COL2])

# 6) 컬럼 싱크 강제 (fit/predict feature mismatch 방지)
Xva = Xva.reindex(columns=Xtr.columns)

# 7) 이후 RF 코드에서 x_train/x_valid 대신 Xtr/Xva를 사용
x_train = Xtr
x_valid = Xva
```

```python

x_train['우편_판매처A'] = x_train['구매지우편번호_te'] * x_train['판매처_A']

x_train['우편_판매처M'] = x_train['구매지우편번호_te'] * x_train['판매처_M']

x_valid['우편_판매처A'] = x_valid['구매지우편번호_te'] * x_valid['판매처_A']

x_valid['우편_판매처M'] = x_valid['구매지우편번호_te'] * x_valid['판매처_M']

x_valid = x_valid.reindex(columns=x_train.columns)


```

```python
x_train['우편_생산국가A'] = x_train['구매지우편번호_te'] * x_train['생산국가_A']

x_train['우편_생산국가TA'] = x_train['구매지우편번호_te'] * x_train['생산국가_TA']

x_valid['우편_생산국가A'] = x_valid['구매지우편번호_te'] * x_valid['생산국가_A']

x_valid['우편_생산국가TA'] = x_valid['구매지우편번호_te'] * x_valid['생산국가_TA']

x_valid = x_valid.reindex(columns=x_train.columns)
```

```python
col_1 =['제조사_b5','차량크기_<NA>','판매처_Other']
x_train = x_train.drop(columns=col_1)
x_valid = x_valid.drop(columns=col_1)
x_valid = x_valid.reindex(columns=x_train.columns)

x_train
```

```python

# 1) 결측치 처리(중앙값) - train 기준으로!
medians = x_train.median(numeric_only=True)

x_train2 = x_train.fillna(medians)
x_valid2 = x_valid.fillna(medians)

# 2) 모델 학습
rf = RandomForestClassifier(
    n_estimators=400,
    random_state=42,
    class_weight="balanced",
    # class_weight=None,
    n_jobs=-1,
    max_depth = 10
)
rf.fit(x_train2, y_train)

# 3) 평가
proba = rf.predict_proba(x_valid2)[:, 1]
pred = (proba >= 0.5).astype(int)

print("ROC-AUC:", roc_auc_score(y_valid, proba))
print("F1:", f1_score(y_valid, pred))
print("\nConfusion Matrix:\n", confusion_matrix(y_valid, pred))
print("\nReport:\n", classification_report(y_valid, pred, digits=4))

# 4) 피처 중요도
import pandas as pd
imp = pd.Series(rf.feature_importances_, index=x_train2.columns).sort_values(ascending=False)
print(imp)
```
