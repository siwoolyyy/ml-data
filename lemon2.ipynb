**train/valid split 후, binary encoding / mean encoding**

```python
train_categories =  x_train['제조사'].unique()
```

```python
# 제조사 값 별로 정수 치환

COL = '제조사'

train_cats = x_train[COL].astype("string").str.strip().unique()

train_cat = pd.Categorical(x_train[COL].astype("string").str.strip(), categories=train_cats)
valid_cat = pd.Categorical(x_valid[COL].astype("string").str.strip(), categories=train_cats)


train_codes = train_cat.codes
valid_codes = valid_cat.codes
```

```python
# 정수를 바이너리 코드로 변경

import numpy as np
import pandas as pd

COL = "제조사"  # 원본 컬럼명

# 1) 비트 수 계산 (제조사 고유값 개수 기준)
n_unique = len(train_cats)   # 너가 이미 만든 train_cats 사용
n_bits = int(np.ceil(np.log2(n_unique)))  # 필요한 비트 수

# 2) train_codes / valid_codes -> 바이너리 컬럼 생성
train_bin = pd.DataFrame(
    {f"{COL}_b{i}": ((train_codes >> i) & 1) for i in range(n_bits)},
    index=x_train.index
)

valid_bin = pd.DataFrame(
    {f"{COL}_b{i}": ((valid_codes >> i) & 1) for i in range(n_bits)},
    index=x_valid.index
)

# 3) 원본 컬럼 제거 + 바이너리 컬럼 붙이기
x_train = pd.concat([x_train.drop(columns=[COL]), train_bin], axis=1)
x_valid = pd.concat([x_valid.drop(columns=[COL]), valid_bin], axis=1)
```


**차량크기-one-hot Encoding**

```python
import pandas as pd

COL3 = "차량크기"

def norm(s):
    return (
        s.astype("string")
         .str.strip()
         .str.replace(r"\s+", " ", regex=True)
    )

train_cats3 = norm(x_train[COL3]).unique()

x_train_cat3 = pd.Categorical(norm(x_train[COL3]), categories=train_cats3)
x_valid_cat3 = pd.Categorical(norm(x_valid[COL3]), categories=train_cats3)

train_oh = pd.get_dummies(x_train_cat3, prefix=COL3, dummy_na=True).astype("uint8")
valid_oh = pd.get_dummies(x_valid_cat3, prefix=COL3, dummy_na=True).astype("uint8")

# ★ 인덱스 정렬 핵심
train_oh.index = x_train.index
valid_oh.index = x_valid.index

x_train = pd.concat([x_train.drop(columns=[COL3]), train_oh], axis=1)
x_valid = pd.concat([x_valid.drop(columns=[COL3]), valid_oh], axis=1)

x_valid = x_valid.reindex(columns=x_train.columns, fill_value=0)

assert x_train.shape[0] == y_train.shape[0]
assert x_valid.shape[0] == y_valid.shape[0]
```

**구매지우편번호-Mean Encoding**

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import StratifiedKFold

TARGET = "lemon"   # 참고용 (여기서는 y_train을 따로 받으니 직접 쓰진 않음)
COL2 = "구매지우편번호"  # TODO: 실제 컬럼명으로 바꿔

def norm_zip(s: pd.Series) -> pd.Series:
    return s.astype("string").str.strip()

def mean_encode_oof_train_only(
    x_train: pd.DataFrame,
    y_train: pd.Series,
    col: str,
    m: float = 20.0,
    n_splits: int = 5,
    seed: int = 42
) -> pd.Series:
    """
    train에 대해 OOF mean encoding 값을 생성해서 Series로 반환.
    누수 방지: 각 행의 인코딩은 그 행이 속한 fold를 제외한 데이터로만 계산.
    """
    X = x_train
    y = y_train

    X[col] = norm_zip(X[col])
    global_mean = y.mean()

    oof = np.zeros(len(X), dtype=float)
    skf = StratifiedKFold(n_splits=n_splits, shuffle=True, random_state=seed)

    for fit_idx, hold_idx in skf.split(X, y):
        X_fit = X.iloc[fit_idx]
        y_fit = y.iloc[fit_idx]

        stats = y_fit.groupby(X_fit[col]).agg(["sum", "count"])
        enc_map = (stats["sum"] + m * global_mean) / (stats["count"] + m)

        oof[hold_idx] = (
            X.iloc[hold_idx][col]
            .map(enc_map)
            .fillna(global_mean)
            .astype(float)
            .to_numpy()
        )

    return pd.Series(oof, index=x_train.index, name=f"{col}_te")
```
