베이즈 정리


```python
#베이즈정리 함수 정의
#pA : 사전확률 (1공급업자로부터 또는 2공급업자로부터 공급받는 확률)
#pB : 전체 상품 중 Good(정상품)일 확률. 또는 Bad (불량품)일 확률
#pB_A : 1공급업자로부터 또는 2공급업자로부터 공급받은 상품 중에 Bad (불량품)일 확률

def bayes(pA, pB, pB_A):
    return(pA * pB_A) / pB
```

```python
# 1단계 함수 정의 (사전확률)

pA1 = 0.65
pA2 = 0.35

# 2단게 조건부 확률 정의
pG_A1 = 0.98
pB_A1 = 1 - pG_A1

pG_A2 = 0.95
pB_A2 = 1 - pG_A2
```

```python
# pG : 정상품일 전체 확률  / pB : 불량품일 전체 확률 
```

```python
pG = pA1 * pG_A1 + pA2 * pG_A2
pB = pA1 * pB_A1 + pA2 * pB_A2
```

```python
# 베이즈 정리 적용
pA1_G = bayes(pA1, pG, pG_A1)
pA2_G = bayes(pA2, pG, pG_A2)
pA1_B = bayes(pA1, pB, pB_A1)
pA2_B = bayes(pA2, pB, pB_A2)
```

```python
# 결과 출력

print(f"P(A1/G) = {pA1_G:.4f}")
print(f"P(A2/G) = {pA2_G:.4f}")
print(f"P(A1/B) = {pA1_B:.4f}")
print(f"P(A2/B) = {pA2_B:.4f}")
```
