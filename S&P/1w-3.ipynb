```python
#사용할 컬럼만 가져오기
df = pd.read_excel(r'C:\Users\siwoo\OneDrive\바탕 화면\데이터관련자료\통확실습코드\Restaurant.xlsx', usecols=[1,2])
df
```

```python
conditions = [
                (df['Meal Price ($)']>=10)&(df['Meal Price ($)']<20),
                (df['Meal Price ($)']>=20)&(df['Meal Price ($)']<30),
                (df['Meal Price ($)']>=30)&(df['Meal Price ($)']<40),
                (df['Meal Price ($)']>=40)&(df['Meal Price ($)']<50),
]

options = ['10~19', '20~29', '30~39', '40~49']

df['Group'] = np.select(conditions, options, default='Not Specified')

df
```

```python
crosstable = pd.crosstab(index=df['Quality Rating'], columns=df['Group'])
crosstable
```

```python
# 행으로 합계
crosstable['Sum'] = crosstable.sum(axis=1)

crosstable
```

```python
# 열로 합계
crosstable.loc['Sum'] = crosstable.sum(axis=0)

crosstable
```
