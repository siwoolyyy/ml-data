```python
# 정렬 (sorting)
sorted_salary = np.sort(salary)
sorted_salary
```

```python
# 중앙값 구하기
n = len(sorted_salary)

if n % 2 == 0:
    x0 = sorted_salary[n//2 - 1]
    x1 = sorted_salary[n//2]
    median = (x0 + x1) / 2

else:
    median = sorted_salary[(n+1)//2 - 1]

float(median)
```

```python
# 중앙값 구하기 -2

np.median(salary)

float(np.median(salary))
```

```python
# 최빈값

salary.mode()
```

```python
# 분산 
# 모집단 분산 vs 표본 분산

# 모집단 분산 : 편차 제곱을 합계 ==> 모집단 요소의 갯수로 나눠줌.

# 표본 분산: 편차 제곱을 합계 ==> 표본 크기에서 1을 뺀 값, 즉 자유도로 나눠줌.

# ddof : delta degree of freedom
# ddof : 자유도를 n-1로 한다.
```

```python
np.var(salary)
```

```python
np.var(salary, ddof=1)
```

```python
np.var(salary) * 12 / 11
```

```python
# 표준 편차

np.std(salary)
```

```python
float(np.sqrt(np.var(salary)))
```

```python
np.std(salary, ddof=1)
```

```python
salary.var(ddof=1)
```

```python
salary.std(ddof=1)
```
