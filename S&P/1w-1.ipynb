```python
brands = df['Brand Purchased']

brands
```

```python
brands.unique()
```

```python
# brands.unique()
# series만 되는 pandas.series.unique

brand, counts = np.unique(brands, return_counts=True)

# 이게 dataframe 됨 numpy array.
# return_counts=True -> 갯수까지 세줌.

brand, counts
```

```python
frequency = brands.value_counts()
frequency
```

```python
n = len(brands)
n
```

```python
# 상대도수
rel_freq = frequency / n
rel_freq
```

```python
pd.DataFrame({'Brand': brand, 'Frequency': frequency, 'Relative Frequency': rel_freq})
```

```python
pd.DataFrame({'Frequency': frequency, 'Relative Frequency': rel_freq})
```

```python
plt.bar(brand,counts)

#그래프 이름
plt.title('Brand Purchased')

#x축 이름
plt.xlabel('Brand')

#y축 이름
plt.ylabel('Frequency')


plt.show()
```

```python
plt.pie(counts, labels=brand, autopct='%.1f%%')

#그래프 이름
plt.title('Brand Purchased')

plt.show()
```
