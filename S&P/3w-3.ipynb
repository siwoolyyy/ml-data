```python
import numpy as np
import pandas as pd
import scipy.stats as stats
import math
```

```python
# 소수 4자리까지 출력
%precision 4 
np.set_printoptions(precision=4, suppress=True) # numpy 출력
pd.options.display.float_format = '{:.4f}'.format # pandas 출력
```

```python
# 이항분포
```

```python
n = 3
p = 0.3

exp = n * p
var = n * p * (1 - p)
std = np.sqrt(var)

exp, var, std 
```

```python
# 마틴 옷가게 문제
```

```python
x_set = np.arange(0,4)
x_set
```

```python
for x in x_set:
    prob = stats.binom.pmf(x, n=3, p=0.3) # pmf : probability mass function 확률질량함수
    print(f"p({x})={prob.round(4)}")
```

```python
# 누적분포함수 cumulative distribution function CDF

for x in x_set:
    cum_prob = stats.binom.cdf(x, n=3, p=0.3)
    print(f"P(X<={x})={cum_prob.round(4)}")

```

## 포아송분포

```python
# 교재에서는 평균을 mu로 사용, 우리 실습에서는 Lambda 사용 (6장 지수분포에서 혼선 방지)
```

```python
def poisson(x, lam):
    return (lam ** x) * np.exp(-lam) / math.factorial(x)
```

```python
poisson(5,10).round(4)
```

```python
lam = 10
x_set = np.arange(0,11)

for x in x_set:
    prob = stats.poisson.pmf(x, lam)
    print(f"p({x}={prob:.4f}")
```

```python
for x in x_set:
    cum_prob = stats.poisson.pmf(x, lam)
    print(f"p(X<={x}={cum_prob:.4f}")
```

**hyper geometric distribution**

```python
def comb(n,x):
    return math.factorial(n) / (math.factorial(x) * math.factorial(n-x))
```

```python
# 전체갯수 n / 골라낸갯수 m / 전체성공횟수 r / 골라낸 성공갯수 x (변수)
def hyper_prob(x, n, r, m):
    return comb(r, x) * comb(n-r, m-x) / comb(n, m) 
```

```python
n = 12   # 박스 안의 모든 퓨즈 갯수
r = 5    # 박스 안의 모든 불량품 갯수
m = 3    # 박스에서 12개 퓨즈 중에 골라내는 표본의 갯수 (검사대상)
x = 1    # 표본에 들어있는 불량품 갯수

type(hyper_prob(x, n, r, m))
```
