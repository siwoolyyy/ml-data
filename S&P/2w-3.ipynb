**조건부확률**

```python
data = {'Male': [288, 672], 'Female': [36, 204]}
df = pd.DataFrame(data=data)
df
```

```python
df1 = pd.DataFrame(data=data, index=["Promote", "Non-Promote"])
df1
```

```python
# 컬럼
df1['Sum'] = df1['Male'] + df1['Female'] 

# 인덱스
df1.loc['Sum'] = [np.sum(df1.iloc[:,0]), np.sum(df1.iloc[:,1]), np.sum(df1.iloc[:,2])]

cross_tab = df1
cross_tab
```

```python
col_percent = cross_tab.iloc[:,:2] / cross_tab.iloc[2,:2]
col_percent
```

```python
joint_prob_tab = cross_tab / cross_tab.loc['Sum', 'Sum']
joint_prob_tab
```

```python
col_percent2 = joint_prob_tab.iloc[:, :3] / joint_prob_tab.iloc[2,:3]
col_percent2
```
