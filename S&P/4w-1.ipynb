```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
from scipy.integrate import quad
```

6.1 균일확률분포

```python
def f(x):
    x_range = np.array([120,140])
    if x_range[0] <= x <= x_range[1]:
        return 1 / (140 - 120)
    else:
        return 0
```

```python
a = 120
b = 140

mean = (a+b)/2
var = (b-a) ** 2 / 12  

mean, var
```

```python
round((b-a) ** 2 / 12, 4) 
```

```python
def P(alpha, beta):
    return quad(f, alpha, beta)[0] #tuple이기 때문에 첫 번째 값 추출
```

```python

P(120, 130)
```

```python
# 누적 확률 

def cum_P(x):
    return stats.uniform.cdf(x, loc=a, scale=b-a)
```

```python
cum_P(130) 
```

```python
#확률밀도함수 그래프
x = np.linspace(a-10, b+10, 1000)

pdf_values = stats.uniform.pdf(x, loc=a, scale=b-a)

plt.plot(x, pdf_values, label='Uniform Distribution')
plt.title('Uniform Distribution (120~140)')
plt.xlabel('values')
plt.ylabel('pdf')
plt.legend()
plt.show()
```

```python
# 125 ~ 130 누적확률계산

cum_P(130) - cum_P(125)
```
