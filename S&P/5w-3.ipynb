8.1 모집단 평균의 추정 (모집단 분산이 알려진 경우)

```python
lloyds = pd.read_excel('lloyds.xlsx')
lloyds
```

```python
lloyds.head()
```

```python
lloyds.tail()
```

```python
rv = stats.norm(0,1)
```

```python
n = len(lloyds['Amount Spent'])
n
```

```python
sample_mean = lloyds['Amount Spent'].mean()
sample_mean
```

```python
# 모집단 표준편차가 주어짐
std = 20
```

```python
# 95% 신뢰구간
# 오른쪽 z값 = 오른쪽 면적이(1-0.95)/2 = 0.025이 되는 x축의 값
# 통계량 (x축의 값) 왼쪽 면적 : cdf (누적분포함수)
# 통계량 (x축의 값) 오른쪽 면적 : sf (생존함수)
# 왼쪽 면적의 경계선이 되는 통계량 : ppf (percent point function)
# 오른쪽 면적의 경계선이 되는 통계량 : isf (inverse survival function)

rv.isf(0.025).round(2)
```

```python
se_sample_mean = std / np.sqrt(n)
se_sample_mean
```

```python
# 오차한계 (E)
e = rv.isf(0.025) * se_sample_mean
e
```

```python
# 95% 신뢰구간의 상한 upper confidence limit
# 95% 신뢰구간의 하한 lower confidence limit

lcl = sample_mean - e
ucl = sample_mean + e

lcl.round(2), ucl.round(2)
```
