8.2 모집단 평균의 추정 (모집단 분산이 알려지지 않은 경우)

```python
# 표본표준편차 s로 대신하고, student t 분포

import numpy as np
import pandas as pd
import scipy.stats as stats
```

```python
new = pd.read_excel('newbalance.xlsx')
new
```

```python
n = len(new['NewBalance'])
n
```

```python
rv_t = stats.t(n-1)
```

```python
sample_mean = new['NewBalance'].mean()
sample_std = new['NewBalance'].std(ddof=1)
se_sample_mean = sample_std / np.sqrt(n)

# 표본평균, 표본표준편차, 평균의 표준오차
sample_mean, sample_std, se_sample_mean
```

```python
# 95% 신뢰구간의 오른쪽 경계선 t값
# (1 - 0.95) / 2 = 0.025

rv_t.isf(0.025).round(4)
```

```python
# 오차한계 (E)
e = rv_t.isf(0.025) * se_sample_mean
e
```

```python
# 95% 신뢰구간의 상한 upper confidence limit
# 95% 신뢰구간의 하한 lower confidence limit

lcl = sample_mean - e
ucl = sample_mean + e

lcl.round(0), ucl.round(0)
```
