```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
```

6.4 지수확률분포

```python
# 361쪽 누적확률
# 교재에서 지수분포함수에 사용하는 평균 (mu)은 포아송분포에서 사용하는 평균 lamda의 역수 1/lamda, mu = 1/lamda 
def exp_prob(lam, x):
    return 1 - np.exp(-lam * x)
```

```python
mean = 10              # 1대분 선적에 소요되는 평균 시간 10분
lam = 1/mean         # 1분에 선적할 수 있는 평균 횟수 1/10회
```

```python
# 그림 6.17
# 1대분 선적에 6분~12분 소요될 확률
# 12분 소요될 확률 - 6분 소요될 확률
exp_prob(lam, 12), exp_prob(lam,6)
```

```python
# (exp_prob(lam, 12) - exp_prob(lam,6)).round(4).item()
# np.exp(-lam * 6) - np.exp(-lam * 12)

(np.exp(-lam * 6) - np.exp(-lam * 12)).round(4).item()
```

```python
x = np.linspace(0, 30, 1000)
pdf_values = stats.expon.pdf(x, scale=mean)

plt.plot(x, pdf_values, label='Exponential PDF')
plt.title('Exponential Distribution')
plt.xlabel('values')
plt.ylabel('pdf')
plt.legend()
plt.show()
```

```python
# 1대분 선적에 6분~12분 소요될 확률
# exp_prob(lam, 12), exp_prob(lam,6)

(stats.expon.cdf(12, scale=mean) - stats.expon.cdf(6, scale=mean)).round(4).item()
```
