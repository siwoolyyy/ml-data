## 그룹화


```python
audit_group = np.where(audit<=14, '14일 이하',  
                        np.where(audit<=19, '19일 이하',
                                 np.where(audit<=24, '24일 이하',
                                          np.where(audit<=29, '29일 이하',
                                                   '30일 이상'))))
audit_group 
```

```python
# audit_counts = pd.Series(audit_group).value_counts
# audit_counts
```

```python
audit_counts = pd.Series(audit_group).value_counts()
audit_counts
```

```python
# 정렬 결과 원본 객체에 바로 반영 inplace=False가 Default 값
audit_counts.sort_index(inplace=True)

audit_counts
```
