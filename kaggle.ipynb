```python
from sklearn.model_selection import RandomizedSearchCV, train_test_split
from sklearn.metrics import roc_auc_score
import numpy as np
from scipy.stats import randint, uniform

# 데이터
feature_cols = [col for col in df1.columns if col not in ['SEQ', '매칭성공여부']]
x = df1[feature_cols].fillna(0)
y = df1['매칭성공여부']
x_train, x_val, y_train, y_val = train_test_split(x, y, test_size=0.2, random_state=42, stratify=y)

# 1. XGBoost RandomizedSearch (30회 샘플링, 5분)
from xgboost import XGBClassifier
xgb_param_dist = {
    'n_estimators': randint(200, 600),
    'max_depth': randint(4, 10),
    'learning_rate': uniform(0.05, 0.15),
    'reg_alpha': uniform(1, 5),
    'reg_lambda': uniform(1, 5)
}
xgb_rs = RandomizedSearchCV(
    XGBClassifier(enable_categorical=True, random_state=42),
    xgb_param_dist, n_iter=30, cv=3, scoring='roc_auc', n_jobs=-1, random_state=42, verbose=1
)
xgb_rs.fit(x_train, y_train)
print(f"XGB 최적: {xgb_rs.best_score_:.4f}, 파라: {xgb_rs.best_params_}")

# 2. CatBoost
from catboost import CatBoostClassifier
cat_param_dist = {
    'iterations': randint(300, 700),
    'depth': randint(4, 8),
    'learning_rate': uniform(0.05, 0.15),
    'l2_leaf_reg': uniform(3, 7)
}
cat_rs = RandomizedSearchCV(
    CatBoostClassifier(random_seed=42, verbose=0),
    cat_param_dist, n_iter=20, cv=3, scoring='roc_auc', n_jobs=-1, random_state=42
)
cat_rs.fit(x_train, y_train)
print(f"Cat 최적: {cat_rs.best_score_:.4f}, 파라: {cat_rs.best_params_}")

# 3. RF
from sklearn.ensemble import RandomForestClassifier
rf_param_dist = {
    'n_estimators': randint(200, 500),
    'max_depth': [8, 12, 16, None],
    'min_samples_split': randint(2, 20)
}
rf_rs = RandomizedSearchCV(
    RandomForestClassifier(random_state=42, n_jobs=-1),
    rf_param_dist, n_iter=25, cv=3, scoring='roc_auc', n_jobs=-1, random_state=42
)
rf_rs.fit(x_train, y_train)
print(f"RF 최적: {rf_rs.best_score_:.4f}, 파라: {rf_rs.best_params_}")

# 최종 앙상블 (개선된 모델들)
xgb_best = xgb_rs.best_estimator_
cat_best = cat_rs.best_estimator_
rf_best = rf_rs.best_estimator_

xgb_proba = xgb_best.predict_proba(x_val)[:,1]
cat_proba = cat_best.predict_proba(x_val)[:,1]
rf_proba = rf_best.predict_proba(x_val)[:,1]

# 가중 최적화
best_auc = 0
best_w = (0.4, 0.35, 0.25)
for xgb_w in np.arange(0.35, 0.5, 0.05):
    for cat_w in np.arange(0.3, 0.4, 0.05):
        rf_w = 1 - xgb_w - cat_w
        ens = xgb_proba*xgb_w + cat_proba*cat_w + rf_proba*rf_w
        auc = roc_auc_score(y_val, ens)
        if auc > best_auc:
            best_auc = auc
            best_w = (xgb_w, cat_w, rf_w)
print(f"최종 앙상블: {best_auc:.4f}")

# testset 제출
seq_values = testset['SEQ'].astype(str) if 'SEQ' in testset.columns else testset.index.astype(str)
x_test = testset[feature_cols].fillna(0).reindex(columns=feature_cols, fill_value=0)

ensemble_test = (xgb_best.predict_proba(x_test)[:,1] * best_w[0] + 
                 cat_best.predict_proba(x_test)[:,1] * best_w[1] + 
                 rf_best.predict_proba(x_test)[:,1] * best_w[2])

submission = pd.DataFrame({'SEQ': seq_values, 'pred': ensemble_test}).sort_values('SEQ').reset_index(drop=True)
submission.to_csv('rs_3model_final_75.csv', index=False)
print("🎯 RandomizedSearch 앙상블 제출 완료!")
```
